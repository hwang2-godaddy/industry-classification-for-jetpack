```python
from pyspark.sql import SparkSession

spark = SparkSession\
    .builder\
    .appName("Spark Session in Local")\
    .getOrCreate()
```

```python
input_df = spark.read.csv("query3_results.csv", header=True, inferSchema=True)
input_df.count()
```

```python
input_df.show()
```

```python
input_df.createOrReplaceTempView("input_df")
```

```python
distinct_label = spark.sql(
    """
    select
        business_category
    from input_df
    where business_category is not null
    group by 1
    order by 1
    """
)
```

```python
distinct_label.show()
```

```python
distinct_label.count()
```

```python
label_list = distinct_label.select("business_category").rdd.flatMap(lambda x: x).collect()

values = [i for i in range(len(label_list))]
```

```python
label_list
```

```python
values
```

```python
import pandas as pd

LABEL_825_DF = pd.DataFrame(list(zip(label_list, values)), columns =['business_category', 'label_int'])

LABEL_825_DF.iteritems = LABEL_825_DF.items

LABEL_825_DF_spark = spark.createDataFrame(LABEL_825_DF)
LABEL_825_DF_spark.createOrReplaceTempView("LABEL_825_DF_spark")
```

```python
LABEL_825_DF.to_csv("LABEL_825_DF.csv")
```

```python
LABEL_825_DF_spark.show()
```

```python
input_df_v2 = spark.sql(
    """
    select
        A.domain_name as feature,
        B.label_int as label
    from input_df A
    left join LABEL_825_DF_spark B
    on A.business_category = B.business_category
    where A.business_category is not null
    """
)
```

```python
input_df_v2.show()
```

```python
input_df_v2.createOrReplaceTempView("input_df_v2")
```

```python
temp_df = spark.sql(
    """
    select * from input_df_v2
    order by rand()
    limit 10
    """
).show()
```

```python
temp_df = spark.sql(
    """
    select 
        count(*)
    from input_df_v2
    where label is null
    """
).show()
```

```python
temp_df = spark.sql(
    """
    select 
        label, 
        count(*) as total_counts
    from input_df_v2
    group by 1
    order by total_counts desc
    """
).show(1000)
```

```python
# save as parquet

input_df_v2.repartition(1).write.parquet(f"training_dataset.parquet",mode="overwrite")
```
